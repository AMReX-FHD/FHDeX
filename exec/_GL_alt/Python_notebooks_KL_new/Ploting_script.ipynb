```python
# This script can be used to plot the contigous umrella data in "AVG_DATA.txt" output (contains ALL umbrella data)
# AVG_DATA.txt has 4 columns with the following data:
# phi_averages, GL energy( no umbrella term), GL energy with umbrella, H1 seminorm

# This is a convienient script to run while producing data
# In particular it is useful for checking that sufficient overlap occured
# at places where the data "jummps" since tuning of parameters can be nuanced




import sys, os, os.path
import glob

import scipy as sp
import numpy as np

import matplotlib
import matplotlib.pyplot as pp
from mpl_toolkits.mplot3d import Axes3D

import yt
from yt.frontends.boxlib.data_structures import AMReXDataset

from tempfile import TemporaryFile


%pylab inline
```

```python
dat = np.loadtxt('AVG_DATA.txt')
A=dat.shape
num_umb=int(A[0]/int(200)) # divide by the number of samples collected in each umbrella
num_umb #number of umbrellas used
```

```python
plt.plot(dat[:,0],'or',markersize=0.1)
pp.ylabel('$\phi_x$')
pp.xlabel('$n \quad (\,index \,\, of \, \quad t_n)$')
pp.title('Spatial Average of $\phi$')
```

```python
plt.plot(dat[:,3],'or',markersize=.1)
pp.ylabel('$|\phi|_{H^1}$')
pp.xlabel('$n \quad (\,index \,\, of \, \quad t_n)$')
pp.title('$H^1$ seminorm')
```

```python
plt.plot(dat[:,2],'or',markersize=.1)
pp.ylabel('$F[\phi]$')
pp.xlabel('$n \quad (\,index \,\, of \, \quad t_n)$')
pp.title('GL Free Energy Functional with umbrella')
```
