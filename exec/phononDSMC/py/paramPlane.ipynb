```python
#number of surfaces
noOfParamPlanes = 40;
radius = 0.015;


theta = 0.0;
delt = 2.0*math.pi/noOfParamPlanes;

offset = [0.02,0.02,0.0];

#making a cylinder out of 40 rectangles
for ii in range(noOfParamPlanes):
    #define parametric surface
    x0=math.cos(ii*delt)*radius+offset[0];
    y0=math.sin(ii*delt)*radius+offset[1];
    z0 = 0.0; 
    
    x1=math.cos((ii+1)*delt)*radius+offset[0];
    y1=math.sin((ii+1)*delt)*radius+offset[1];
    z1=0.0;
    
    #tangent and parameter 1
    uTop = np.linalg.norm([x1-x0,y1-y0,z1-z0],2)
    u=[x1-x0,y1-y0,z1-z0]/uTop;
    ux=u[0]; uy=u[1]; uz=u[2];
    
    #tangent and parameter 2
    vx=0.0; vy=0.0; vz=1.0;
    vTop = heightZ;
    
    #surface normals
    xt=math.cos((ii+0.5)*delt)*radius+offset[0];
    yt=math.sin((ii+0.5)*delt)*radius+offset[1];
    zt=heightZ/2;
    ln=[offset[0]-xt,offset[1]-yt,heightZ/2-zt]/np.linalg.norm([offset[0]-xt,offset[1]-yt,heightZ/2-zt],2);
    lnx=ln[0]; lny=ln[1]; lnz=ln[2];
    rnx=-lnx; rny=-lny; rnz=-lnz;
    
    #porosity, specularity, temperature, record particles (left and right)
    pl=0.0; sl=2e-7; tl=3.0;cl=0.0;
    pr=0.0; sr=2e-7; tr=3.0;cr=0.0;
    
    #periodic
    p=0.0;
    
    #density if a source, for each species (left and right)
    nr0=0.0; nr1=0.0;nl0=0.0;nl1=0.0;
    #source
    sourcer=0.0;sourcel=0.0;
    #area (only used if a source)
    a=0.0;
    #sink
    sinkl=0.0;
    sinkr=0.0;
    
    #code expects it in this order:
    paramPlane=[x0,y0,z0,ux,uy,uz,vx,vy,vz,uTop,vTop,rnx,rny,rnz,lnx,lny,lnz,pr,sr,tr,nr0,nr1,sourcer,sinkr,cr,pl,sl,tl,nl0,nl1,sourcel,sinkl,cl,p,a];
    
    paramPlaneList.append(paramPlane);
    
```

```python
thetaList = [];
for ii in range(len(velList)):
    xVel = velList[ii][0]
    zVel = velList[ii][2]
    theta = np.arctan(zVel/xVel)
    thetaList.append(theta)
```

```python
nb=80
hist=np.zeros(nb)
frac = math.pi/nb

for ii in range(len(thetaList)):
    theta = thetaList[ii] + math.pi/2
    bin = math.floor(theta/frac)
    hist[bin] = hist[bin] + 1
    
halffrac = frac/2
tVal=np.zeros(nb)
for ii in range(nb):
    tVal[ii] = ii*frac + halffrac - math.pi/2
    hist[ii] = hist[ii]/len(thetaList)
```
