# Test Implementation of Marker Placement on Sphere

## Load Data

This data is populated by configuring the `main3d` program to output marker positions.

```python
f = open("sphere_markers.txt")
a = readlines(f)
close(f)
```

Each line (after line 5) contains the positions of of the markers, represented as $(x, y, z)$ tuples:

```python
a[end-1:end]
```

### Parse Data:

```python
n_markers = length(a[6:end])
marker_positions = Array{Float64}(undef, (3, n_markers))

index = 1;
for line=a[6:end]
    
    # parse line:
    #  * `[2:end-1]` <= to strip the leading and trailing `()`
    a_line::Array{SubString{String}} = split(line[2:end-1], ",");
    marker_positions[:, index] = [parse(Float64, elt) for elt = a_line]
    index += 1
end
```

```python
marker_positions
```

```python
plot3D(marker_positions[1, :], marker_positions[2, :], marker_positions[3, :], "o")

# Adjust "viewing angle"
gca()[:view_init](20,30)

# Equal and fixed axes
xlim([.3,.5]);
ylim([.3,.5]);
zlim([.3,.5]);
```
