```python
import yt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def return_y_dens_cplt(folder,time,nbins,field):
    ds = yt.load(folder+'/pltcu'+time+'/')
    ad = ds.all_data()
    plot = yt.slicePlot(ad,"z",[field],weight_field="cell_volume",x_log=False,y_log={field:False})
    profile = plot.profiles[0]
    x = np.array(profile.x)
    y = np.array(profile[field])
    return x, y
```
